```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('IPL.csv')

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python

for i, col in enumerate(df.columns):
    print(f"{i+1}. {col}")
```

```python
# Top 10 batsmen by total runs
top_batsmen = df.groupby('batter')['runs_batter'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette='Oranges_r')
plt.title('Top 10 IPL Run Scorers (2008-2025)', fontsize=16, fontweight='bold')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.tight_layout()
plt.show()
```

```python
wickets_df= df[df['bowler_wicket'] == True]

top_bowlers = wickets_df.groupby('bowler')['bowler_wicket'].count().sort_values(ascending=False).head(10)

plt.Figure(figsize=(12,6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='Blues_r')
plt.title('Top 10 IPL Wicket Trackers (2008-2025)', fontsize=16, fontweight='bold')
plt.xlabel('Total Wickets')
plt.ylabel("Bowler")
plt.tight_layout()
plt.show()
```

```python

df['year']= pd.to_datetime(df['date']).dt.year


season_runs= df.groupby('year')['runs_total'].sum()


plt.figure(figsize=(12, 6))
plt.plot(season_runs.index,season_runs.values,
         marker='o', color='green',linewidth=2.5)
plt.fill_between(season_runs.index, season_runs.values,
                 alpha=0.3, color='green')
plt.title('Season-wise Total Runs in IPL',fontsize=16,fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Runs')
plt.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()
```

```python

team_runs= df.groupby('batting_team')['runs_total'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=team_runs.values, y=team_runs.index, palette='viridis')
plt.title('Total Runs Scored by Each IPL Team', fontsize=16, fontweight='bold')
plt.xlabel('Total Runs')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```
